### Step 9. Which was the most-ordered item?

```python
c = chipo.groupby('item_name').sum()
c = c.sort_values(['quantity'], ascending=False)
c.head(1)
```

### Step 10. For the most-ordered item, how many items were ordered?

```python
c = chipo.groupby('item_name').sum()
c = c.sort_values(['quantity'], ascending=False)
c.drop(columns = ['order_id' , 'choice_description' , 'item_price'] , inplace=True)
c.head(1)
```

### Step 11. What was the most ordered item in the choice_description column?

```python
c = chipo.groupby('choice_description').sum()
c = c.sort_values(['quantity'],ascending=False)
c.drop(columns = ['order_id', 'item_price', 'item_name'] , inplace=True)
c.head(1)
```

### Step 12. How many items were orderd in total?

### Step 13. Turn the item price into a float

#### Step 13.a. Check the item price type

```python
chipo.item_price.dtype
```

#### Step 13.b. Create a lambda function and change the type of item price

```python
dollarizer = lambda x: float(x[1:-1]) if isinstance(x, str) else x
chipo.item_price = chipo.item_price.apply(dollarizer)
```

#### Step 13.c. Check the item price type

```python
chipo.item_price.dtype
```

### Step 14. How much was the revenue for the period in the dataset?

```python
revenue = (chipo['quantity']* chipo['item_price']).sum()
print('Revenue was: $' + str(np.round(revenue,2)))
```

### Step 15. How many orders were made in the period?

```python
order = chipo.order_id.value_counts().count()
order
```

### Step 16. What is the average revenue amount per order?

```python
chipo['revenue'] = chipo['quantity'] * chipo['item_price']
order_grouped = chipo.groupby(by=['order_id'])['revenue'].sum()
order_grouped.mean()
```
