```python
import pandas as pd
import sklearn
```

### Step 1. Load the data

```python
 url = 'https://raw.githubusercontent.com/HassanAlgoz/B5/main/content/datasets/Rounded_Student_Hours_Studied_vs_Marks_Dataset.csv'
 data = pd.read_csv(url)
```

```python
data.head(5)
```

### Step 2.a Assign variables `X` to the features and `y` to the target

```python
X = data['Hours_Studied']
y = data['Marks']
```

### Step 2.b print the type of each

```python
print(type(X))
print(type(y))
```

### Step 2.c identify whether the target is categorical or numerical & whether the task is regression or classification

- The target variable (Marks) is numerical, as it represents continuous values (student scores).

- Therefore, the problem is a regression task, since we are predicting a numeric outcome rather than class labels.

### Step 3. Identify the number of samples and columns of both the data matrix and the target

```python
print("X shape:", X.shape)
print("y shape:", y.shape)
```

### Step 4. Summarize the distribution of the data (min, max, median, mean, and standard deviation)

```python
X.describe()
```

### Step 5. Plot the feature vs the target

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(X, y, alpha=0.7)
plt.title('Hours Studied vs Marks')
plt.xlabel('Hours Studied (Feature)')
plt.ylabel('Marks (Target)')
plt.grid(True)
plt.show()
```

### Step 6. What is the relationship between the feature and the target? (increasing or decreasing or none)

increasing

### Step 7. Initialize a `LinearRegression` model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

### Step 8. Split the dataset into train and test sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X , y)
```

### Step 9.a Train the model on the training set

```python
model.fit(X_train.values.reshape(-1, 1), y_train)
```

### Step 9.b Identify the learned slope and the y-intercept of the model

```python
print("Slope : " , model.coef_)
print("y-intercept : " ,model.intercept_)
```

### Step 9.c how much marks do we get per 1 hour of study?

```python
model.coef_
```

### Step 9.d how much does a 10 minutes increase of study time adds to our marks?

```python
marks_increase = model.coef_[0] * (10/60)
print('Marks increased: ' ,  marks_increase)
```

### Step 9.e if one doesn't study at all, how much marks do they get?

```python
model.intercept_
```

- The intercept of the model represents the predicted marks when Hours_Studied is 0.

- Therefore, if a student does not study at all, they are expected to score approximately (model.intercept_) marks.

### Step 10. Evaluate the model on the test set

```python
from sklearn.metrics import mean_squared_error

predictes = model.predict(X_test.values.reshape(-1,1))
mse = mean_squared_error(y_test, predictes)
print("MSE : " , mse)
```

### Step 11. Take a sample of three rows from the test set and make a prediction on them and check whether they match the true label or not

```python
import pandas as pd

X_sample = X_test.iloc[:3]
y_true = y_test.iloc[:3]

# Reshape X_sample to be 2-dimensional (n_samples, 1_feature)
y_pred = model.predict(X_sample.values.reshape(-1, 1))

comparison = pd.DataFrame({
    "Hours_Studied": X_sample.squeeze(),  # convert to Series for cleaner view
    "Actual Marks": y_true,
    "Predicted Marks": y_pred
})

print(comparison)
```

### Step 12. Plot the data and the regression line

```python
import matplotlib.pyplot as plt
import numpy as np

# Scatter plot (actual data)
plt.scatter(X, y, color='blue', label='Actual Data')

# Create smooth range for line
X_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)

# Predict values for the line
y_line = model.predict(X_range)

# Plot regression line
plt.plot(X_range, y_line, color='red', label='Regression Line')

# Labels and title
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.title("Hours Studied vs Marks with Regression Line")
plt.legend()

plt.show()
```
