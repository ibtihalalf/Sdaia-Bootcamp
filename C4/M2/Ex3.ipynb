```python
from docling.document_converter import DocumentConverter, PdfFormatOption
from docling.datamodel.pipeline_options import PdfPipelineOptions
from docling.datamodel.base_models import InputFormat

# 1) Create pipeline options
pipeline_options = PdfPipelineOptions()

# Required for picture classification
pipeline_options.generate_picture_images = True
pipeline_options.images_scale = 2

# Enable Picture Classification enrichment
pipeline_options.do_picture_classification = True

# 2) Create converter with custom PDF pipeline
converter = DocumentConverter(
    format_options={
        InputFormat.PDF: PdfFormatOption(
            pipeline_options=pipeline_options
        )
    }
)

# 3) Convert document
result = converter.convert("https://arxiv.org/pdf/2501.17887")

# 4) Get Docling document
doc = result.document

print("Conversion completed with Picture Classification enrichment.")
```

```python
# Print picture classifications
for item, _ in doc.iterate_items():
    if item.label == "picture":
        print(item)
```
